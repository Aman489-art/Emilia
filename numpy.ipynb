**lEXSORT**

```python
#lexsort, which is an indirect stable sort on multiple keys,

a = [1, 5, 1, 4, 3, 4, 4]  # First sequence

b = [9, 4, 0, 4, 0, 2, 1]  # Second sequence

ind = np.lexsort((b, a))  # Sort by `a`, then by `b`
ind
[(a[i], b[i]) for i in ind]
```

a = primary

b = secondary

STEP 1 -  first sort primary key

| Value | Indexes |
| ----- | ------- |
| 1     | 0, 2    |
| 3     | 4       |
| 4     | 3, 5, 6 |
| 5     | 1       |

[0, 2, 4, 3, 5, 6, 1]


STEP 2 - SORT SAME A GROUPS ACCORDING TO B GROUPS

| Index | a | b |
| ----- | - | - |
| 0     | 1 | 9 |
| 2     | 1 | 0 |
| 3     | 4 | 4 |
| 5     | 4 | 2 |
| 6     | 4 | 1 |


FINAL:
[2, 0, 4, 6, 5, 3, 1]

STEP 3 :
[(a[i], b[i]) for i in ind]

[(1, 0), (1, 9), (3, 0), (4, 1), (4, 2), (4, 4), (5, 4)]

# Partition

np.partition() rearranges the elements of an array so that:

the kth smallest element is placed in its correct position (as it would appear in a fully sorted array)

all smaller elements come before it

all larger elements come after it

but the left and right parts are not fully sorted
(only separated)

```python
a = np.array([7, 1, 5, 3, 9, 2])
np.partition(a, 2)
```

```python
np.partition(a, 2, axis=-1, kind='introselect', order=None)
```

# a → the array

# kth → which position to fix

# axis → direction of partition (for multi-dimensional arrays)
**axis=0** → column-wise

**axis=1** → row-wise

**axis=-1** (default) → last axis

# kind='introselect'`
Algorithm used internally.

# order
Used only for structured arrays (arrays with named fields).
Normally you keep it None.

# convert a 1D array into a 2D array

Using np.newaxis will increase the dimensions of your array by one dimension when used once. This means that a 1D array will become a 2D array, a 2D array will become a 3D array, and so on.

```python

#np.newaixs

a = np.array([1, 2, 3, 4, 5, 6])
a.shape

#to add new axis
a2 = a[np.newaxis, :]
a2.shape
#to convert 1d array in a row vector
ro_vector = a[np.newaxis, :]
ro_vector.shape
#to convert 1d array into column vector
col_vector = a[:, np.newaxis]
col_vector.shape

```

```python
#np.expand_dims
#add an axis at index position 1
b = np.expand_dims(a, axis=1)
b
```

```python
#add an axis at index position 0
b = np.expand_dims(a, axis=0)
b
```
